```python
def extract_tar(tar_path, debug = False):
    # Check if the tar file exists
    if not os.path.exists(tar_path):
        print(f"Error: The file '{tar_path}' does not exist.")
        return

    # Extract the tar file
    with tarfile.open(tar_path, 'r') as tar:
        tar.extractall()
        extracted_to = tar.getmembers()[0].name
    if debug:
        print(f"Extraction complete. Files are extracted to '{extracted_to}'")
    return extracted_to
```

```python
def extract_graph_from_stp(file_path):
    graph = nx.Graph()
    reading_edges = False
    node_count = 0
    i = 0
    with open(file_path, 'r') as file:
        for line in file:
            if i == 0:
                node_count = int(line)
                for j in range(1, int(node_count)+1):
                    graph.add_node(j)
            if i > 1:
                node1, node2, weight = line.split()
                graph.add_edge(int(node1), int(node2), weight=int(weight))
            i += 1
    return graph

def process_stp_files_in_folder(folder_path, file_extension, file_names = None, debug = False):
    graphs = {}
    file_names = list(file_names)
    
    # Loop through all files in the given folder
    for filename in os.listdir(folder_path):
        process = True
        if filename.endswith(file_extension):
            if file_names is not None:
                if not filename in file_names:
                    process = False
            if process:
                file_path = os.path.join(folder_path, filename)

                if debug:
                    print(f"Processing file: {filename}")

                graph = extract_graph_from_stp(file_path)
                graphs[filename] = graph

                if debug:
                    print(f"Graph has {len(graph.nodes())} nodes and {len(graph.edges())} edges")
                
    return graphs
```

```python
def get_graph_data(file_path, file_extension):
    # Read the Excel file
    df = pd.read_excel(file_path, usecols=['Instanz', 'K', 'ZF'])

    # Change K from number of edges to number of nodes
    df['K'] = df['K'] + 1

    # Normalize file names
    df['Instanz'] = df['Instanz'].str[:-len(file_extension)]
    df['Instanz'] += file_extension
    df = df.dropna()
    df['2_approximation'] = 0
    df['Calculated'] = False
    df['Time'] = '0:00:00.00'
    df['Ratio'] = 0
    df.rename(columns={'Instanz': 'File_Name', 'ZF': 'True_Weight'}, inplace = True)
    
    return df
```

```python
def format_time(seconds):
    # Calculate hours, minutes, and seconds
    hours = int(seconds // 3600)
    minutes = int((seconds % 3600) // 60)
    seconds = seconds % 60  # Remaining seconds
    
    # Format the time string
    time_str = f"{hours:02}:{minutes:02}:{seconds:05.2f}"
    
    return time_str
```

```python
def plot_ratios(df, graph_file_name, graphs, folder_name = None):
    for file_name in df['File_Name']:
        df.loc[df['File_Name'] == file_name, 'Nodes'] = len(graphs[file_name].nodes)
    
    df['K_Percentage'] = df['K'] / df['Nodes']
    for file_name, group in df.groupby('File_Name'):
        plt.plot(group['K_Percentage'], group['Ratio'], marker='o', linestyle='-', label=file_name)

        #plt.title(f'Weight Ratios for {graph_file_name}')
    plt.xlabel('k / |V|')
    plt.ylabel('Ratio')
    plt.legend()
    plt.ylim(.97, 1.52)
    
    if folder_name:
        folder_path = os.path.join('results', folder_name)
        if not os.path.exists(folder_path):
            os.makedirs(folder_path)
        full_path = os.path.join(folder_path, graph_file_name + '.png')
    else:
        full_path = graph_file_name + '.png'
    plt.savefig(full_path, dpi=300)
    plt.show()
    
plot_ratios(bluble_regular, 'bluble_regular', regular_graphs, 'ratio_plots')
```
